```python
import pandas as pd
```

Cleaning player data:

```python
player_data = pd.read_csv('data/player_stats_clean.csv')

players = player_data
players.rename(columns={'Name': 'Player'}, inplace=True)
```

Cleaning MVP data:

```python
mvp_voting = pd.read_csv('data/mvp_voting.csv')
```

```python
# remove rows where 'Rk' or 'Age' are null
mvp_voting = mvp_voting.dropna(subset=['Rk'])

# remove Unnamed: 0 column
mvp_voting = mvp_voting.drop('Unnamed: 0', axis=1)

# set 'MVP' to 'True' if Rk is 1
mvp_voting['MVP'] = mvp_voting['Rk'] == 1
```

```python
mvp_voting.head()
```

```python
# Remove Rk column
mvp_voting = mvp_voting.drop('Rk', axis=1)

mvps = mvp_voting[["Player", "Year", "Votes", "Share", "MVP"]]
```

```python
mvps.head()
```

```python
players = players.merge(mvps, on=["Player", "Year"], how="left")

# drop rows where 'MVP' is null
players = players.dropna(subset=['MVP'])
```
